## 유전자형 데이터 & 표현형 데이터 전처리(Munging)

요구하는 데이터?
- 유전체 연구 데이터(GWAS 통계랑 데이터, Genotype-Phenotype 매칭 데이터 등등....)

데이터 형식 & 정합성

- PLINK 형식(.bed, .bim, .fam), CSV 형식 지원!!
- **PLINK**: 유전학 및 생명정보학에서 사용하는 도구로, 유전자형 데이터분석 및 품질 관리, GWAS(Genome-Wide Association Study)같은 대규모 유전체 데이터 전처리 & 분석에 쓰임

PLINK 파일 형식?
1. .bed: 바이너리 형식으로 저장된 유전형 데이터
2. .bim: SNP 정보 포함!!(SNP ID, 염색체 번호, 위치 등등)
3. .fam: 샘플 정보, 표현형 데이터(ID, 성별, 케이스/컨트롤)

```python
!ls outputs/Munge/
```

```python
!genoml discrete supervised munge \
--prefix outputs \
--geno /content/genoml2/examples/discrete/training \
--pheno /content/genoml2/examples/discrete/training_pheno.csv
```
